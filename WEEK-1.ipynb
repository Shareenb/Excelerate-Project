# Week 1 : Data Cleaning and Exploratory Data Analysis

```python
#Loading the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data set
df1 = pd.read_csv("UserData (2).csv")
df2 = pd.read_csv("Opportunity Wise Data.csv")
```

```python
# Get basic statistics for numerical columns
user_stats = df1.describe()

# Count null values for each column
user_null_counts = df1.isnull().sum()

# Display the results
print("\nUserData Statistics:")
print(user_stats)
print("\nUserData Null Values:")
print(user_null_counts)
```

```python
# Get basic statistics for numerical columns
op_stats = df2.describe()

# Count null values for each column
op_null_counts = df2.isnull().sum()

# Display the results
print("\nOpp-Data Statistics:")
print(op_stats)
print("\nOpp-Data Null Values:")
print(op_null_counts)
```
